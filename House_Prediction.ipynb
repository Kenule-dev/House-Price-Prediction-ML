```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

house_price=pd.read_csv('/content/train.csv')

print(house_price)

X=house_price.drop(['BHK_OR_RK', 'ADDRESS', 'POSTED_BY'], axis=1) #Remove the unneeded colums

print(X)


Y=house_price['TARGET(PRICE_IN_LACS)'] # Y value

print('Y is', Y )

X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.30, random_state=42) # Splitting dataset into % for training and testing


#X_test = np.reshape(X_test, (-1, 1))

print('Y_test is:', Y_test)

model=LinearRegression() # Setting up the model

model.fit(X_train, Y_train) # Training your dataset

prediction=model.predict(X_test) # Predicting X with respect to Y

print(f'Predicted is:{prediction}')

accuracy=r2_score(Y_test, prediction) # Test the accuracy of the model

print(f"Accuracy is:{accuracy:.2f}")

#To plot the dataset


plt.scatter(Y_test, prediction, label='Actual Data', color='Red') # Plotting the original data

plt.plot([Y_test.min(), Y_test.max()],[Y_test.min(), Y_test.max()] , label='Predicted value', color='Green', linewidth=3) # Plot of the predicted data

plt.title('House Price in Delhi, India', fontsize=14, fontweight='bold', color='Violet') # Add title

plt.xlabel('Needed criteria', fontsize=14, color='Black') # Add x-axis

plt.ylabel('Price in Lacs', color='Black', fontsize=14) # Add y-axis

plt.legend() # Add legend

plt.grid(True, color='Pink') # Add grid

plt.show()




```
